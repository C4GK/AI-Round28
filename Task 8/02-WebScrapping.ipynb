## Companies Revenue

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Define the URL of the Wikipedia page
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
```

```python
# Send an HTTP GET request to the URL
response = requests.get(url)
html_content = response.content
```

```python
# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# Find the table containing the company data
table = soup.find('table', class_='wikitable')
```

```python
# Extract company names and revenues
companies = []
for row in table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    if len(columns) >= 2:
        company_name = columns[1].get_text(strip=True)
        revenue = columns[2].get_text(strip=True)
        companies.append({'Company': company_name, 'Revenue': revenue})
```

```python
# Print the collected data
for company in companies:
    print(f"Company: {company['Company']}")
    print(f"Revenue: {company['Revenue']}")
    print("----------")
```

## Book Shop

```python
import requests
from bs4 import BeautifulSoup
```

```python
# Define the base URL and starting page number
base_url = 'https://books.toscrape.com/catalogue/page-{}.html'
```

```python
page_number = 1
# Loop through multiple pages
while True:
    url = base_url.format(page_number)

    # Send an HTTP GET request to the URL
    response = requests.get(url)
    if response.status_code != 200:
        break

    html_content = response.content

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all book article elements on the page
    books = soup.find_all('article', class_='product_pod')

    # Loop through each book and extract relevant information
    for book in books:
        title = book.h3.a['title']
        price = book.find('p', class_='price_color').text
        rating = 'No rating' if 'star-rating' not in str(book) else book.find('p', class_='star-rating')['class'][-1]

        print(f"Title: {title}")
        print(f"Price: {price}")
        print(f"Rating: {rating}")
        print("----------")

    # Move to the next page
    page_number += 1
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
# Initialize the WebDriver (ensure you have chromedriver installed and in PATH)
driver = webdriver.Chrome()
```

```python
# Open Amazon website
driver.get("https://www.amazon.com/")
```

```python
# Find the search bar and input the book title
search_bar = driver.find_element(By.ID, "twotabsearchtextbox")
search_bar.send_keys("Python Programming")
```

```python
# Submit the search
search_button = driver.find_element(By.XPATH, "//input[@type='submit']")
search_button.click()
```

```python
# Find book title and price elements
book_titles = driver.find_elements(By.XPATH, "//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']")
book_prices = driver.find_elements(By.XPATH, "//span[@class='a-price']/span[@class='a-offscreen']")
```

```python
# Extract and print book titles and prices
for i in range(len(book_titles)):
    title = book_titles[i].text
    price = book_prices[i].text
    print(f"Title: {title}")
    print(f"Price: {price}")
    print("----------")
```

```python
# Close the browser
driver.quit()
```
